```python
import pandas as pd

file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df = pd.DataFrame()

for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name)

    # データフレームの列名を指定
    temp_df.columns = ['売上年', '商品', '金額（千円）']

    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df], ignore_index=True)

```

```python
sum_goods = (
    df
    .groupby(['商品', '売上年'], as_index=False)['金額（千円）']
    .sum()
)

print(sum_goods)

```

```python
sum_year = df.groupby('売上年',as_index=False)['金額（千円）'].sum()

print(sum_year)
```

```python
writer = pd.ExcelWriter('売上集計表.xlsx')

sum_goods.to_excel(writer, sheet_name='Sheet1', index=False)

writer.close()
```

```python
import openpyxl

from openpyxl.styles import PatternFill

wb = openpyxl.load_workbook('売上集計表.xlsx')
sheet = wb.active

fill = PatternFill(patternType='solid', fgColor='F2F2F2')
for row in sheet['A1:C1']:
    for cell in row:
        cell.fill = fill


wb.save('売上集計表.xlsx')
```
