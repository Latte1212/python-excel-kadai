```python

# 表のヘッダー行を書き込む
header = ['商品名', '数量', '単価', '金額']
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000],
]

all_data = [header] + data

for i, row in enumerate(all_data):
    for j, value in enumerate(row):
        ws.cell(row=10+i, column=2+j, value=value)

wb.save(file_path)
```
